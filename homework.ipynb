# Load the book

```python
with open('miracle_in_the_andes.txt', 'r', encoding='UTF-8') as file:
    book = file.read()
```

# Extract the paragraphs where 'love' was used

```python
import re
```

```python
pattern = re.compile('[^\n]+love[^\n]+')
result = re.findall(pattern, book)
result[:2]
```

# Extract the chapter titles

### Method 1

```python
pattern = re.compile('[a-zA-Z ]+\n\n')
result = re.findall(pattern, book)
result = [item.strip('\n\n') for item in result]
result
```

### Method 2 - only return the part in the () directly

```python
pattern = re.compile('([a-zA-Z ]+)\n\n')
result = re.findall(pattern, book)
result
```

# Function that finds the occurence of any word

```python
def find(w):
    pattern = re.compile('[a-zA-Z]+')
    result = re.findall(pattern, book)
    d = {}
    for word in result:
        if word in d.keys():
            d[word] = d[word] + 1
        else:
            d[word] = 1
    try:
        return d[w]
    except:
        return f'The book does not contain the word "{w}"'
```

# Call the function

```python
find('love')
```

```python
find('hate')
```
